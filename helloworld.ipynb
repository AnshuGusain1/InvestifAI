```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import date



try:
    tickr = input("Please enter the ticker: ").strip().upper()

    start_date = input("Please enter a start date (Format (YYYY-MM-DD)): ").strip()

    end_date = input("Please enter an end date (Format (YYYY-MM-DD)): ").strip()
    if(end_date.strip().lower() == "today"):
        end_date = date.today().strftime("%Y-%m-%d")

    # Fetch historical data
    stock = yf.Ticker(tickr)
    historical_data = stock.history(start=start_date, end=end_date, interval="1d")

    # Plot the data
    plt.plot(historical_data.index, historical_data['High'])
    plt.xlabel('Date')
    plt.ylabel('High Price')
    plt.title(f'{tickr} Highs from {start_date} to {end_date}')
    plt.show()

except Exception as e:
    print(f"An error occurred: {e}")
```
